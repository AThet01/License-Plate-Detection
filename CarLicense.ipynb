```python
import zipfile
import os

# Define file paths
zip_path = "/content/License Plates.v3-original-license-plates.yolov12.zip"  # Replace with your ZIP file path
extract_folder = "unzipped_dataset"  # Folder where files will be extracted

# Create the folder if it doesn't exist
os.makedirs(extract_folder, exist_ok=True)

# Extract ZIP file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

print(f"Files extracted to: {extract_folder}")
```
